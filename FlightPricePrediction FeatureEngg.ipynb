## Featuring engineering 

Here in date_of_journey we have date in a format so we take the feature and we want the to divide feature data into
day,month,year i.e 3 categories in total

```python
# we are in the process to create derived feature on Date_of_Journey data
# we split the data by '/'
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
# considering th feature Arrival
#here the feature arrival_time has time as well as date also so we are selecting only the time for this arrival feature 

final_df['Arrival_Time']=final_df['Arrival_Time'].str.split().str[0]
#also
#final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x:x.split(' ').[0])
```

```python
final_df.isnull().sum()
```

```python
#getting the hour and minute category from Arrival_Time

final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0].astype(int)
final_df['Arrival_minute']=final_df['Arrival_Time'].str.split(':').str[1].astype(int)
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df
```

```python
#getting the hour and minute category from Arrival_Time

final_df['Dep_hour']=final_df['Dep_Time'].str.split(':').str[0].astype(int)
final_df['Dep_minute']=final_df['Dep_Time'].str.split(':').str[1].astype(int)
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
# route feature
#As in the data we have route and the journey flight_details
#in Total_stops we are given with the no of stops between destination and source
# Therefore we can neglet the Riute Column 

# here now we are mapping the Total_stops with respect to no_of_stops
final_df['Total_Stops'].isnull().sum()
```

```python
final_df['Total_Stops'].unique()
```

```python
# mapping the values of total_stops and for nan we are replacing 1 we can give any value based on the situation
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df.head()
```

```python
#dropping the Route feature as no longer needed

```

```python
final_df.drop('Route',axis=1,inplace=True)
```

```python
#for Additional_Info feature
final_df['Additional_Info'].unique()

# here we can use one hot encoded format as we are having less values 
```

```python
#for duration we can convert the total duration in h:m format to mins as in movie tickets 
final_df['Duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
#.str.split(' ').str[0]. we get the 2h as output we write further .str stmt to remove 'h' 
#.str.split(' ').str[0].str.split('h').str[0]
```

```python
#to convert it to int type
final_df['Duration_hour']=final_df['Duration_hour'].astype(int)
# we get the error as we have 5m data in the Duration_hour feature
```

```python
# we drop the records as it is immpossible by analysing for 5m of journey with 2 Total_stops 
final_df.drop(6474,axis=0,inplace=True)     #we use axis=0 as we are deleting row
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Duration_hour']=final_df['Duration_hour'].astype(int)
```

```python
final_df.info()
```

```python
final_df['Duration_min']=final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
final_df['Duration_min'].astype(int)
# error becoz it has nan value 
```

```python
final_df['Duration_min']=final_df['Duration_min'].fillna(0)
```

```python
final_df['Duration_min']=final_df['Duration_min'].astype(int)
```

```python
final_df['Duration_hour']=final_df['Duration_hour']*60
```

```python
final_df['Duration_hour']
```

```python
#we have converted the h and m to mins

final_df['Total_duration']=final_df['Duration_min']+final_df['Duration_hour']
```

```python
final_df.head()
# as we have total_duration lets drop Duration_hour,Duration_min,and Duration
```

```python
final_df.drop('Duration',axis=1,inplace=True)
final_df.drop('Duration_min',axis=1,inplace=True)
final_df.drop('Duration_hour',axis=1,inplace=True)
final_df.head()
```
